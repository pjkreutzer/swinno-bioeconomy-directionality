```python
import polars as pl
from pyhere import here
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Example data
data = pl.read_csv(here("data/raw/bioeconomy_visions.csv"))

agg_data = (
    data.group_by("bioeconomy_vision")
    .agg(pl.col("sinno_id").n_unique().alias("count"))
    .filter(~pl.col("bioeconomy_vision").is_in(["Unsure", "Not Bioeconomy"]))
)
data
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

data = [
    {"bioeconomy_vision": "Biotechnology Vision", "count": 38},
    {"bioeconomy_vision": "Bioecology Vision", "count": 97},
    {"bioeconomy_vision": "Bioresource Vision", "count": 254},
]

# Create a dictionary to map each vision to its count
vision_count = {item["bioeconomy_vision"]: item["count"] for item in data}

# Extract visions and counts
visions = list(vision_count.keys())
counts = list(vision_count.values())

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")

# Set x, y, z coordinates for each point
x_values = [0] * len(visions)  # x-axis corresponds to Biotechnology Vision
y_values = [0] * len(visions)  # y-axis corresponds to Bioecology Vision
z_values = [0] * len(visions)  # z-axis corresponds to Bioresource Vision

# Assign count values to their corresponding axes
for i, vision in enumerate(visions):
    if vision == "Biotechnology Vision":
        x_values[i] = vision_count[vision]
    elif vision == "Bioecology Vision":
        y_values[i] = vision_count[vision]
    elif vision == "Bioresource Vision":
        z_values[i] = vision_count[vision]

# Plot each data point
for i in range(len(visions)):
    ax.scatter(x_values[i], y_values[i], z_values[i], label=visions[i], s=100)

ax.set_xlabel("Biotechnology Vision")
ax.set_ylabel("Bioecology Vision")
ax.set_zlabel("Bioresource Vision")
ax.set_title("Visualization of Aggregated Data")
ax.legend()
plt.show()
```
